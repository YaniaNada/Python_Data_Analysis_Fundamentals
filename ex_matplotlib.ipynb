Ex. 4 : Create a figure with multiple subplots, each showing a different metric (sales, expenses, profit) over time.

```python
# Create a figure with multiple subplots
fig, axs = plt.subplots(3)

# Create a line plot of the sales data
axs[0].plot(data_df['year'], data_df['sales'])
axs[0].set_title('Sales Over Time')

# Create a line plot of the expenses data
axs[1].plot(data_df['year'], data_df['expenses'])
axs[1].set_title('Expenses Over Time')

# Create a line plot of the profit data
axs[2].plot(data_df['year'], data_df['profit'])
axs[2].set_title('Profit Over Time')

# Display the plot
fig.tight_layout()
plt.show()
```

Ex. 5 : Add annotations to the scatter plot to highlight important data_df points or trends, and customize the appearance of the plot using various options, such as colors, markers, line styles, etc.

```python
# create a scatter plot of the sales and profit data
sales_max = data_df['sales'].max()
profit_max = data_df['profit'].max()
sales_min = data_df['sales'].min()
profit_min = data_df['profit'].min()
plt.scatter(data_df['sales'], data_df['profit'], c= 'red')
# Set the title and labels for the plot
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
# Add annotations to highlight important data points or trends
plt.annotate('Highest profit', xy=(sales_max, profit_max), xytext= (sales_max-20000, profit_max-3000), arrowprops=dict(arrowstyle="simple", facecolor='black', shrinkB = 5), horizontalalignment='right', verticalalignment='top')
plt.annotate('Lowest profit', xy=(sales_min, profit_min), xytext = (sales_min+25000, profit_min+5000), arrowprops= dict(arrowstyle="simple", facecolor='black', shrinkB = 5), horizontalalignment ='left', verticalalignment='bottom')
# Display the plot
plt.show()
```
